```python
import pandas as pd
```

```python
combustiveiscanoas = pd.read_excel("combustiveis-2023-02.xlsx")
```

```python
display(combustiveiscanoas[['Endereco', 'Numero', 'Bairro ', 'Cep', 'Municipio', 'Estado', 'Produto', 'Price']])
```

```python
pricegas = combustiveiscanoas[['Endereco', 'Numero', 'Bairro ', 'Cep', 'Municipio', 'Estado', 'Produto', 'Price']]
```

```python
listcombustivelstate = pricegas.loc[ (pricegas['Municipio'] == 'CANOAS') & (pricegas['Estado'] == 'RIO GRANDE DO SUL') & (pricegas['Produto'] == 'GASOLINA COMUM') & (pricegas['Price'] < 5.00)]
display(listcombustivelstate)
```

```python
listcombustivelstate.sort_values(by='Price', inplace=True)
combusteveisCanoas2023 = listcombustivelstate.head(100)
display(combusteveisCanoas2023)
```

```python
listcombustivelstate = pricegas.loc[(pricegas['Estado'] == 'RIO GRANDE DO SUL') & (pricegas['Produto'].isin(['ETANOL','GASOLINA COMUM','GASOLINA ADITIVADA'])) & (pricegas['Price'] < 4.50)]
listcombustivelstate.sort_values(by='Price', inplace=True)
display(listcombustivelstate.head(15))
```

```python
combusteveisCanoas2023.to_excel('combustiveisCanoasFevereiro.xlsx', sheet_name='Çanoas 2023 Fevereiro')
```
