```python
import pandas as pd
```

```python
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
display(combustiveis_df.head())
```

```python
combustiveis_df['Ativo'] = True

display(combustiveis_df.head())
```

```python
combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'CANOAS' else None for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'] == 'CANOAS', ['Municipio', 'Obs']])
```

```python
import numpy as np

combustiveis_df['Status do Valor de Venda'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Status do Valor de Venda']])
```

```python
num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
merge_df.dropna(axis='columns')
print(merge_df.info())
```

```python
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.reset_index(inplace=True)
postos_por_municipio_df.drop(['CNPJ da Revenda', 'Valor de Compra', 'Obs'], axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "NumPostos"}, inplace=True)
display(postos_por_municipio_df)

postos_por_municipio_df['PostosPorHabitanteQDrado'] = postos_por_municipio_df['NumPostos'] / postos_por_municipio_df['NumHabitantes2021']
display(postos_por_municipio_df)
```

```python
postos_por_municipio_df.rename({"Número de Postos": "num postos"}, inplace=True)
postos_por_municipio_df.rename['PostosPorHabitanteQDrado'] = postos_por_municipio_df['num postos'] / postos_por_municipio_df['NumHabitantes2021']
display(postos_por_municipio_df)
```
